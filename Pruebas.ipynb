```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style="whitegrid")

tips = sns.load_dataset("tips")   # numricas + categricas (propina, cuenta, da, fumador, etc.)
penguins = sns.load_dataset("penguins") # numricas + categricas (especie, isla, etc.)
titanic = sns.load_dataset("titanic")   # muchas categricas + numricas

pearson_corr = tips["total_bill"].corr(tips["tip"]) # Calcular correlacin de Pearson

plt.figure(figsize=(10,5))
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='sex', style='time', alpha=.50, s=50)
sns.regplot(data=tips, x='total_bill', y='tip', color='red', scatter=False)
plt.title('scatter Total Bill VS TIP')
plt.xlabel('Total Bill (USD)')
plt.ylabel('TIP (USD)')
plt.figtext(.35, .70,'Pearson Correlation: 0.68', fontsize=12, color='red')
```

```python
pearson_corr = penguins["bill_length_mm"].corr(penguins["flipper_length_mm"]) # Calcular correlacin de Pearson
sns.scatterplot(data=penguins, x='bill_length_mm', y='flipper_length_mm', hue='species', style='sex')
sns.regplot(data=penguins, x='bill_length_mm', y='flipper_length_mm', color='red', scatter=False)
plt.title('Scatter Bill Length VS Flipper Length')
plt.xlabel('Bill Length (mm)')
plt.ylabel('Flipper Length (mm)')

```
